```python
from transformers import EsmModel, EsmTokenizer
import torch
import requests

def get_protein_sequence(ensp_id):
    url = f"https://rest.ensembl.org/sequence/id/{ensp_id}?content-type=text/plain"
    response = requests.get(url)
    
    if response.ok:
        return response.text
    else:
        raise Exception(f"Failed to retrieve sequence for {ensp_id}: {response.status_code}, {response.text}")

# Choose a model (e.g., esm2_t33_650M_UR50D)
model_name = "facebook/esm2_t33_650M_UR50D"

# Load tokenizer and model
tokenizer = EsmTokenizer.from_pretrained(model_name)
model = EsmModel.from_pretrained(model_name)
model.eval()

for ensp_id in gene_list:  # Replace with your ENSP ID
    sequence = get_protein_sequence(ensp_id)

    # Set to eval mode
    inputs = tokenizer(sequence, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)

    # Get per-residue (per-token) embeddings
    token_embeddings = outputs.last_hidden_state  # shape: (1, L+2, hidden_size)

    # Remove [CLS] and [EOS] tokens
    token_embeddings = token_embeddings[0, 1:-1]

    # Mean pooling (excluding [CLS] and [EOS])
    mean_embedding = token_embeddings.mean(dim=0)  # shape: (hidden_size,)

    save the mean_embedding as the protein's feature, in df
save df as csv
```
